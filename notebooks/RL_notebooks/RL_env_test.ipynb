```python
import gym
import highway_env
from stable_baselines3 import PPO
# from sb3_contrib import RecurrentPPO
ABS_PATH = '/home/elios/pighetti/HighwayDRL'

# Visualisation
import sys
sys.path.append(ABS_PATH)
from scripts.utils import record_videos, show_videos
from ipywidgets import interact, widgets
import glob
```

### Choose the environment to exploit the model in

```python
env_id = widgets.Text()
def f(input_env):
    env_id.value = str(input_env)
interact(f, input_env=['dm-env-v0','acc-dm-env-v0','jam-dm-env-v0','overtaken-dm-env-v0','singleO-dm-env-v0','multipleO-dm-env-v0'])
```

### Choose the model

```python
model_id = widgets.Text()
def m(input_model):
    model_id.value = str(input_model)
interact(m, input_model=glob.glob(f"{ABS_PATH}/training_output/models/*.zip"))
```

```python
env = gym.make(env_id.value)
env.configure({
    "screen_width":1500,
    "screen_height":250
})
env = record_videos(env, path=f'{ABS_PATH}/exploitation/videos/{env_id.value}')
print(model_id.value)

model = PPO.load(model_id.value)

for episode in range(10):

    obs, done = env.reset(), False

    while not done:

        action, _ = model.predict(obs, deterministic=True)
        env.render()
        obs, reward, done, info = env.step(int(action))

env.close()
show_videos()
```
