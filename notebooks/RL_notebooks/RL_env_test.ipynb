```python
import gym
import highway_env

from stable_baselines3 import PPO

model = PPO.load("models/ppo_RL_dummy.zip")
env = gym.make("dm-env-v0")
for _ in range(5):
    obs = env.reset()
    done = False
    while not done:
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        env.render()
```
