### Choose the model

```python
model_id = widgets.Text()

def m(input_model):
    model_id.value = str(input_model)
interact(m, input_model=glob.glob(f"{ABS_PATH}/training_output/models/*.zip"))
```

```python
env = gym.make(env_id.value)
env.configure({
            "observation": {
                "type": "Kinematics",
                "vehicles_count": 7
            },
    "simulation_frequency": 15,
    "screen_width": 1500,
    "screen_height": 250
})

model = PPO.load(model_id.value)

crashes = 0
episode_num = 20

for episode in range(episode_num):

    obs, done = env.reset(), False

    while not done:

        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(int(action))
        if(env.vehicle.crashed):
            crashes +=1
        env.render()
    if episode == 10:
        print('episode num:', episode, "number of crashes:", crashes)

env.close()
print('episode num:', episode_num, "number of crashes:", crashes)
```
