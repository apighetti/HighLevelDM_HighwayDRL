### Choose the environment to exploit the model in

```python
env_id = widgets.Text()
def f(input_env):
    env_id.value = str(input_env)
interact(f, input_env=['dm-multi-agent-v0','dm-env-v0','jam-dm-env-v0','overtaken-dm-env-v0','singleO-dm-env-v0','multipleO-dm-env-v0', 'highway-v0'])
```

### Choose the model

```python
model_id = widgets.Text()

def m(input_model):
    model_id.value = str(input_model)
interact(m, input_model=glob.glob(f"{ABS_PATH}/training_output/models/*.zip"))
```

```python
env = gym.make(env_id.value)
model = PPO.load(model_id.value)

env.configure({
    "simulation_frequency": 5,
    "screen_width": 1500,
    "screen_height": 250
})

crashes = 0
position = 0
episode_num = 5

for episode in range(episode_num):
    if episode == 10:
        print('episode num:', episode, "number of crashes:", crashes)

    obs, done = env.reset(), False

    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(int(action))
        # print(obs, obs.shape, type(obs), done, info)
        if(env.vehicle.crashed):
            crashes +=1
        # if(env.position_terminal):
        #     position +=1
        env.render()
        # print('fr:', env.final_reward, 'sr:', env.sparse_reward, 'dr:', env.dense_reward)
    

env.close()
# print("# of posisiton terminals:", position)
# print('episode num:', episode_num, "number of crashes:", crashes)
```
