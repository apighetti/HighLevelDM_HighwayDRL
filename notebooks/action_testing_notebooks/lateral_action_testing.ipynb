```python
import gym
import highway_env
import time

env = gym.make('overtaken-dm-env-v0')
env.configure({
    "manual_control": True,
    "screen_width":1500,
    "screen_height":250,
    "simulation_frequency": 30
    })

env.reset()
done = False
while not done:
    env.step(env.action_space.sample())
    env.render()    # with manual control, these actions are ignored

env.close()
```
