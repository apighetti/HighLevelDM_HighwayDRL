```python
import gym
import highway_env
import time

env = gym.make('dm-env-v0')
env.configure({
    "manual_control": True,
    "screen_width": 1250,
    'lanes_count': 1,
    'vehicles_count': 2

})

env.reset()
done = False
while not done:
    env.step(env.action_space.sample())
    env.render()    # with manual control, these actions are ignored

env.close()
```
