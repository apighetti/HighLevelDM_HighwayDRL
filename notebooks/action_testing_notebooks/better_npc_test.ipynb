```python
import gym
import highway_env

env = gym.make('acc-dm-v0')
env.configure({
    "manual_control": True,
    "screen_width": 1250,
    "screen_height":100
    
})

env.reset()
done = False
while not done:
    env.step(env.action_space.sample())
    env.render()    # with manual control, these actions are ignored
env.close()
```
