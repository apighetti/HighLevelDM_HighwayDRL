```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = '/Users/fornerispighetti/HighwayDRL/highway_env/ACC_data.csv'

data = pd.read_csv(path)

plt.figure(figsize=(15,10))

plt.plot(np.arange(len(data))/20, data['ego_speed']*3.6,label='EGO VEHICLE')
plt.plot(np.arange(len(data))/20,data['front_vehicle_speed']*3.6, label='FRONT VEHICLE')
plt.grid()
plt.xticks(range(0,350,10))
plt.legend()
```
