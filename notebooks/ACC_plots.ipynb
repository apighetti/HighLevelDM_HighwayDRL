```python
path = '/Users/fornerispighetti/HighwayDRL/highway_env/ACC_data.csv'

data = pd.read_csv(path)
m = 0.006334268445885509


plt.figure(figsize=(25,10))

plt.plot(np.arange(len(data['counter']))*m, data['ego_speed']*3.6,label='EGO VEHICLE')

#
plt.plot(np.arange(len(data['counter']))*m,data['front_vehicle_speed']*3.6, label='FRONT VEHICLE')
plt.xlabel("Time [s]")
plt.ylabel("Speed [km/h]")#
plt.title("Comparison between ego vehicle speed and front vehicle speed when activating adaptive cruise control")
# plt.axvline(x=68.8, color='red', linestyle='dotted')
plt.grid()

plt.legend()
```

```python
plt.figure(figsize=(25,10))

plt.plot(np.arange(len(data['counter']))*m, data['gap'])
plt.xlabel("Time [s]")
plt.ylabel("Time gap [s]")
plt.title("Time headway trend over time when activating adaptive cruise control")
# plt.axvline(x=68.8, color='red', linestyle='dotted')
plt.grid()
```

```python
plt.figure(figsize=(25,10))

plt.plot(np.arange(len(data['counter']))*m, data['throttle'])
plt.xlabel("Time [s]")
plt.ylabel("acceleration [m/s\u00b2]")
plt.title("Ego vehicle throttle trend over time when activating adaptive cruise control")
# plt.axvline(x=68.8, color='red', linestyle='dotted')
plt.grid()
```

```python
plt.figure(figsize=(25,10))

plt.plot(np.arange(len(data['counter']))*m, data['distance'])
plt.xlabel("Time [s]")
plt.ylabel("Distance with front vehicle [m]")
plt.title("Distance between ego vehicle and front vehicle over time when activating adaptive cruise control")
# plt.axvline(x=68.8, color='red', linestyle='dotted')§
plt.xticks(range(0,160,5))
plt.grid()
```
