# Manual control

```python
import gym
import highway_env
import time
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

env = gym.make('dm-env-v0')
env.configure({
    "screen_width": 1500,
    "screen_height": 250,
    "manual_control": True
    })

for i in range(3):
    env.reset()
    done = False
    
    while not done:
        _, _, done, _ = env.step(env.action_space.sample())
        env.render()
        
    env.close()
```
