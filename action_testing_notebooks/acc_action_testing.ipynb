```python
import gym
import highway_env
import time

env = gym.make('acc-dm-v0')
env.configure({
    "manual_control": True,
    "simulation_frequency": 80,
    "screen_width": 1250
    })
env.reset()
done = False
while not done:
    obs,_,_,_ = env.step(env.action_space.sample())
    throttle_obs = obs.observer_vehicle.throttle
    env.render()    # with manual control, these actions are ignored
```
